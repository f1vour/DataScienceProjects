```python
import pandas as pd
import numpy as np


path ="/content/drive/MyDrive/DataSets/company_sales_data.csv"
df = pd.read_csv(path, encoding = 'unicode_escape')

#Exercise 1:  Explore the data:
# Print the first few rows using df.head().
# df aka DataFrame

df.head()
```

```python
# Check the data types of each column using df.info().

# Check data types
print("Data Types and Info:")
df.info()
```

```python
# Get summary statistics (mean, median, standard deviation, etc.) using df.describe().

# Statistics summary
print("Statistics Summary:")
df.describe()
```

```python
# Excerscise 2: Calculate basic statistics:

# Calculate the mean, median, mode, and standard deviation for each numerical feature.
df.fillna(0, inplace=True)

# Mean
mean_values = df.mean()
print(f"Mean Values: {mean_values}")

# Median
median_values = df.median()
print(f"\nMedian Values: {median_values}\n")

# Mode
mode_values = df.mode().iloc[5]
print(f"\nMode Values: {mode_values}\n")

# Standard Deviation
std_values = df.std()
print(f"\nStandard Deviation: {std_values}")
```

```python
# Calculate the correlation between different features using df.corr().

correlation_matrix = df.corr()
print(f"Correlation Matrix: {correlation_matrix}")
```
