```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


path ="/content/drive/MyDrive/DataSets/company_sales_data.csv"
df = pd.read_csv(path, encoding = 'unicode_escape')

# Create a basic line plot

plt.figure(figsize=(10, 5))

# Line plot of total profit over months
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle='-', color='b', label='Total Profit')

# Labels and title
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit Over Months')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot with Matplotlib

plt.figure(figsize=(8, 5))

# Scatter plot
plt.scatter(df['bathingsoap'], df['facewash'], color='r', alpha=0.6)

# Scatter plot between Bathing Soap & Face Wash Sales
# Labels and title
plt.xlabel('Bathing Soap Sales')
plt.ylabel('Facewash Sales')
plt.title('Bathing Soap vs. Facewash Sales')

plt.show()
```

```python
# Visualize a dataset with histograms
plt.figure(figsize=(8, 5))

# Histogram of the total profits
sns.histplot(df['total_profit'], bins=10, kde=True, color='purple')

# Labels and title
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Distribution of Total Profit')

plt.show()
```

```python
# Visualize box plots using Seaborn
plt.figure(figsize=(8, 5))

# Box plot of Total Profit
sns.boxplot(y=df['total_profit'], color='orange')

plt.title('Box Plot of Total Profit')

# Showing the box plot using seaborn
plt.show()
```
